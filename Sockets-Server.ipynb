```python
import socket
import hashlib
import time
```

```python
host = '127.0.0.1'
port = 5000
tcp = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
orig = (host, port)

tcp.bind(orig)
tcp.listen(1)

with open('password.txt', 'r') as arq:
    list_password = []
    for line in arq:
        list_password.append(line[:-1])

hash = "2130"
hash = hashlib.sha1(hash.encode()).hexdigest()

pass_sha1 = []
for i in range(len(list_password)):
        pass_sha1.append(((hashlib.sha1(list_password[i].encode()).hexdigest())+';'))

cont = 0

val = True
while val:
    
    con, cliente = tcp.accept()
    print('\n#######################################################')
    print("Conexão ativa realizada com:", cliente)
    
    con.send(hash.encode()) #envia hash para o cliente
    print("Hashs enviados para:", cliente)
    print("Tamanho do banco de hashs enviado:", len(pass_sha1))
    
    cont+=1
    print("Conexões com o Server:", cont)
    
    val1 = True
    while val1: #envia lista de hashs para o cliente processar
        for i in range(len(pass_sha1)):
            con.sendto(pass_sha1[i].encode(), cliente)
        
        msg = con.recv(1024)
        print(msg.decode())
        
        val1 = False
        
    print('Finalizando conexão do cliente:', cliente)
    print('_______________________________________________________')
    time.sleep(1)
    con.close()
```
